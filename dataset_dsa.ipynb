```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import seaborn as sns
import re
```

```python
df = pd.read_csv("./dataset/cleaned_train.csv")
```

```python
print(min(df.price))
print(max(df.price))
print(df.price.idxmax())
print(len(df.price))
```

```python
plt.figure(figsize=(10, 5))
plt.bar(range(len(df['price'])), df['price'])
plt.xlabel('Item')
plt.ylabel('Price')
plt.title('Price vs Item')
plt.show()
```

#### Apply Logarithmic Scaling to the Pricing Column so it is easier to understand 

```python
df['log_price'] = np.log(df['price'] + 1)
print(min(df.log_price))
print(max(df.log_price))
print(df.log_price.idxmax())

plt.figure(figsize = (10, 5))
plt.bar(range(len(df['log_price'])), df['log_price'])
plt.xlabel('Item')
plt.ylabel('Log(Price + 1)')
plt.title('Log-Scaled Price vs Item')
plt.show()
```
